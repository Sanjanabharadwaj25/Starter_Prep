## Chapter 2

* The weights play a crucial role in determining the output of the classifier. By adjusting the weights during the training process, the model learns to assign higher importance to certain feature based on their contribution to the prediction. 

* A bias is an additional parameter added to each neuron in a neural network. The bias is independent of the input data and serves as an offset or baseline value that helps the model better fit the data and make predictions.

* Learning rate is a hyperparameter in machine learning and deep learning algorithms that controls the step size at which a model updates its parameters (weights and biases) during the training process. It determines how much the model adjusts its parameters in response to the calculated gradients (partial derivatives) of the loss function with respect to those parameters.

* one epoch is completed when all training samples have been used once to update the model's parameters (weights and biases).

Model to get correct weights to predict the Y value for linear regression. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
np.random.seed(42) # Same random numbers are generated every time.
X = 2 * np.random.rand(100, 1) # Generate an array of 100 elements
y = 4 + 3 * X + np.random.randn(100, 1) # Adding noise to prevent overfitting

# Initialize random weight and bias
weight = np.random.randn()
bias = np.random.randn()
print("Starting weight:" ,weight)
print("Starting bias:", bias)

# Learning rate and number of epochs for gradient descent
learning_rate = 0.01
epochs = 100

# Training the model using Gradient Descent
for epoch in range(epochs):
    # Forward pass: Compute the predicted output
    y_pred = X * weight + bias
    
    # Compute the mean squared error loss
    loss = np.mean((y_pred - y)**2)
    print("The loss function for epoch", epoch ,"is ",loss)
    # Backpropogation
    # Backward pass: Compute gradients of loss w.r.t. weight and bias
    gradient_weight = np.mean(2 * (y_pred - y) * X) # Multiply by 2 to simplify the calculation
    gradient_bias = np.mean(2 * (y_pred - y))
    
    # Update the weight and bias using Gradient Descent
    weight -= learning_rate * gradient_weight
    bias -= learning_rate * gradient_bias
    
    #Subtract the weight and bias so that it minimizes the loss function.The update rule for gradient descent can be given as
    #w = w - learning_rate * (dL/dw)
    
    
    # Plot graphs for every 10th iteration
    if(epoch%10==0):
        plt.scatter(X, y)
        plt.plot(X, y_pred, color='red')
        plt.xlabel('X')
        plt.ylabel('y')
        plt.title('Linear Regression')
        plt.show()
    
# Print the final learned weight and bias
print("Learned Weight: ", weight)
print("Learned Bias: ", bias)



```

Giving epoch = 200, may cause the data to be overfitted. 

Early stopping helps to monitor the model's performance on a validation set during training and stop the training process when the performance starts to deteriorate.
